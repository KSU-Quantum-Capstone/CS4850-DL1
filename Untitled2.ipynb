```python
circ = QuantumCircuit(8, 8)

#circ.h(4)
circ.x(5)
circ.x(6)
#circ.h(7)

circ.ccx(4,5,3)

circ.x(4)
circ.x(5)

circ.ccx(4,5,0)

circ.x(4)

circ.ccx(4,5,1)

circ.x(4)
circ.x(5)

circ.ccx(4,5,2)

circ.x(4)

#circ.measure(0,0)
#circ.measure(1,1)
#circ.measure(2,2)
#circ.measure(3,3)
#circ.measure(4,4)
#circ.measure(5,5)

circ.barrier()

circ.reset(4)
circ.reset(5)

#and circuit)
circ.ccx(6,7,4)
circ.ccx(3,4,5)
circ.reset(4)

circ.barrier()

#or circuit)
circ.cx(6,4)
circ.cx(7,4)
circ.ccx(6,7,4)
circ.ccx(2,4,5)
circ.reset(4)

circ.barrier()

circ.x(6)
circ.ccx(1,6,5)
circ.barrier()

circ.measure_all(add_bits=False)



display(circ.draw('mpl'))
```

```python
# aquires the specified backend, transpiles 'circ' for the toppology of the specified backend
# creates a job and adds it to that queue to be ran 100 times
backend = myprovider.get_backend('ibmq_qasm_simulator')
instructions = transpile(circ, backend)
simulation = backend.run(instructions, shots=100)
```

```python
# obtains the results of the 100 runs and plots the histogram of all of the output configurations
results = simulation.result()
counts = results.get_counts(instructions)
display(plot_histogram(counts, figsize = (12,5)))
```
